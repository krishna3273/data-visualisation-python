```python
import seaborn as sb
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Matplotlib & Seaborn default

```python
def sinplot(flip = 1):
   x = np.linspace(0, 14, 100)
   for i in range(1, 5): 
      plt.plot(x, np.sin(x + i * .5) * (7 - i) * flip)
```

```python
#matplotlib example
sinplot()
plt.show()
```

```python
#seaborn default
import seaborn as sb
sb.set()
sinplot()
plt.show()
```

## Available Themes
- Darkgrid
- Whitegrid
- Dark 
- White
- Ticks

```python
themes=['darkgrid',"whitegrid","dark","white","ticks"]
current_theme=np.random.choice(themes)
print(current_theme)
import seaborn as sb
sb.set_style(current_theme)
sinplot()
plt.show()
```

## Removing Axes Spines

```python
#Available  only in white and ticks mode
sb.set_style("ticks")
sinplot()
sb.despine()
plt.show()
```

## Overriding the defaults

```python
print(sb.axes_style())
```

```python
sb.set_style("white",{'axes.edgecolor': 'green'})
sinplot()
sb.despine()
plt.show()
```

## Scaling
### Available Contexts
- Paper

- Poster

```python
contexts=['paper',"notebook","talk","poster"]
current_context=np.random.choice(contexts)
print(current_context)
sb.set_style("white")
sinplot()
sb.despine()
sb.set_context(current_context)
plt.show()
```
